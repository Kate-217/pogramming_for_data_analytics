2. Regression plot tips against the total bill.

```python
# loading the data:
data = sns.load_dataset('tips')
data.head(2)
```

Columns in data:<br>
* total_bill: The total bill (cost) of the meal, including tax.
* tip: The tip given by the customer.
* sex: The gender of the person paying the bill ("Male" or "Female").
* smoker: Whether there were smokers in the party ("Yes" or "No").
* day: The day of the week the meal was served (e.g., "Sun" for Sunday).
* time: The time of day ("Lunch" or "Dinner").
* size: The number of people in the dining party.

```python
#regression plot tips against the total bill
sns.lmplot(x='tip', y='total_bill',data=data, order=1)
plt.title('Linear Regression of Total Bill vs. Tips')
plt.show()
```

Make plot using discrete values

```python
# Number of people (size) vs tip

sns.lmplot(x='tip', y='size',data=data, x_jitter=0.05, x_estimator=np.mean)
plt.title('Linear Regression of Number of peope vs. Tips')
plt.show()
```
